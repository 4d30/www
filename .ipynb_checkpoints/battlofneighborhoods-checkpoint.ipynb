## Introduction: Business Problem <a name="introduction"></a>
In this project, optimal locations for a **diner** in **Brooklyn, NY** will be located.

We will be targetting the 'hipster' demograhic. The definition of the word hipster is amorphous and is outside the scope of this exploration. Neighborhoods will be evaluated upon the quantity and distance to the following types of businesses:
* Organic Grocer
* Bicycle Shop
* Flea Market
* Indie Movie Theatre

We will **minimize the distance to hipster establishments**.

Our diner will have to compete with the existing industry in Brooklyn. The following types of business are our primary competition.
* Diners
* Breakfast Spots
* Coffee Shops
* Cafes
* Burger Joints
* Sandwich Places

We will **maximize the distance to the competition**.

Finally, we will attempt to place our diner near the following major transportation infrastructure:
* The Brooklyn Bridge
* The Manhattan Bridge
* The Williamsburg Bridge

We will **minimize the distance to major transportation infrastructure**.

Data science skills will be used to produce a list of a 10 locations that meet these critera. Each proposed location will be clearly described in terms of the findings of this exploration to assist stakeholders in making the best possible decision in final location.

Now a function is defined to return a numpy array of coordinates when supplied the name of a city and state. First, the filesystem is checked. If the data is not found there a request is sent to OSM.

```python
def get_coords(city, state):
    file = '{}-{}.npy'.format(city,state)
    if os.path.isfile(file):
        print("Found file:\n\t{}".format(file))
        print("Loading data from file...")
        return np.load(file)
    else:
        url = 'https://nominatim.openstreetmap.org/search/{}%20{}?format=json&addressdetails=1&limit=1'.format(city,state)
        print("Trying to fetch coordinates from: \n{}".format(url))
        try:
            response = requests.get(url,headers=headers) 
        except requests.exceptions.RequestException as e:
            raise e
        if response.status_code == 200:
            lat = float(response.json()[0]['lat'])
            lng = float(response.json()[0]['lon'])
            np.save(file,np.array([lat,lng]))
            return np.array([lat,lng])
        else:
            print("Warning: HTTP status {}".format(response.status_code))        
```

#### Grid Generation - *Some light algebra*  <a name="grid-algebra"></a>
Now that we've located the region of interest, let's generate a grid of locations, or "neighborhoods." Our neighborhoods will be located with a pair of coordinates and assigned a regular radius of **RADIUS** meters. Neighborhoods will be hexagonally tessellatted, or placed on a grid of hexagons, to minimize overlap and maximize accuracy.


The **Haversine** equation calculates the length **d** of the great-arc between to points on a sphere or radius **r**, given their latitudes **$\phi_n$** and longitudes **$\lambda_n$**.
<br/><br/>

$d = H(\lambda,\phi,r)$


$d = 2 r \arcsin\left(\sqrt{\sin^2\left(\frac{\phi_2 - \phi_1}{2}\right) + \cos(\phi_1) \cos(\phi_2)\sin^2\left(\frac{\lambda_2 - \lambda_1}{2}\right)}\right)$




```python
r = 6369106.0 # Approx. radius of Earth at 40deg North, 0m above sea-level

def haversine(coords_f, coords_i):
    coords_fr = coords_f*np.pi/180 # Converting to radians
    coords_ir = coords_i*np.pi/180
    a=np.power(np.sin((coords_fr[0] - coords_ir[0])/2),2)
    b=np.cos(coords_fr[0])*np.cos(coords_ir[0])*np.power(np.sin((coords_fr[1]-coords_ir[1])/2),2)
    d=2*r*np.arcsin(np.sqrt(a+b))
    return d
```

We'll need to manipulate this a little bit. What we have is the coordinates for Brooklyn and a distance between neighborhoods of 600m. The neighborhood 600 meters directly north of the the point we requested from OSM has the same longitude and a different latitude. Let's solve for latitude.

<center>
$\frac{d}{2 r} = \arcsin\left(\sqrt{\sin^2\left(\frac{\phi_2 - \phi_1}{2}\right) + \cos(\phi_1) \cos(\phi_2)\sin^2\left(\frac{0}{2}\right)}\right)$
<br/><br/>
<center>
$\sin^2\frac{d}{2 r} = \sin^2\frac{\phi_2-\phi_1}{2}$
<br/><br/>
<center>
$\boxed{\phi_2 = \frac{d + r\phi_1}{r}}$



```python
def lat_f(lat_i, d):
    lat_rads = lat_i*np.pi/180 
    return (d+r*lat_rads)/r *180/np.pi
```

Now we will repeat the procedure. This time latitude is held constant while solving for longitude:
<br/><br/>
<center>
$\frac{d}{2 r} = \arcsin\left(\sqrt{\sin^2\left(\frac{ 0 }{2}\right) + \cos^2(\phi_1)\sin^2\left(\frac{\lambda_2 - \lambda_1}{2}\right)}\right)$
<br/><br/>
<center>
$\sin^2\left(\frac{d}{2 r}\right) = \cos^2\left(\phi_1\right)\sin^2\left(\frac{\lambda_2 - \lambda_1}{2}\right)$
<br/><br/>
<center>
$\sin\left(\frac{\lambda_2 - \lambda_1}{2}\right) = \sin\left(\frac{d}{2 r}\right)\sec\left(\phi_1\right)$
<br/><br/>
<center>
$\boxed{\lambda_2 = 2\arcsin\left(\sin\left(\frac{d}{2 r}\right)\sec\left(\phi_1\right)\right) + \lambda_1}$

```python
def lng_f(lng_i, lat_i, d):
    lat_rads = float(lat_i)*np.pi/180  # Converting to radians
    lng_rads = float(lng_i)*np.pi/180  
    lambda_2 = 2*np.arcsin(np.sin(d/(2*r))/np.cos(lat_rads))+lng_rads # numpy doesn't include secant, cos^-1 has taken its place.
    return lambda_2*180/np.pi
```

#### Grid Generation   <a name="grid-loop"></a>
The coordinates of the  grid of hexagons is produced below. This procedure starts at the southern most point of the central meridian. A pair of coordinates is created before the process moves east until no longer in our search area. This is then repeated due west. Afterwards we move north two hexagonal tiers on the central meridian to repeat the search east and west previously described.

Finally, the entire procedure is repeated with an offset to find neighborhoods in the hexagonal tiers skipped in the first pass.

```python
radius_bk = 8500 # Radius of search area
radius_hd = 600 # Radius of each neighborhood within search ares
df = pd.DataFrame(columns=['coords'])

# Calculate southern point in search region and initialze search.
begin = np.array([lat_f(coords_bk[0],-radius_bk+0.001),coords_bk[1]])
print("BEGIN: {}".format(begin))
loc = begin
df = df.append({"coords":loc},ignore_index=True)
i=0
while haversine(loc,coords_bk) < radius_bk:
    loc_new = np.array([lat_f(loc[0],2*np.sqrt(3)*radius_hd),loc[1]]) # Move north
    loc = loc_new
    bookmark = loc # setting a bookmark so we can return here later
    if haversine(loc,coords_bk) > radius_bk:
        break
    df = df.append({"coords":loc},ignore_index=True)
    while True: # Move east
        loc_new = np.array([loc[0],lng_f(loc[1],loc[0],2*radius_hd)])  
        loc = loc_new
        if haversine(loc,coords_bk) > radius_bk:
            break
        else:
            df = df.append({"coords":loc},ignore_index=True)
    loc = bookmark
    while True: # Move west
        loc_new = np.array([loc[0],lng_f(loc[1],loc[0],-2*radius_hd)])  
        loc = loc_new
        if haversine(loc,coords_bk) > radius_bk:
            break
        else:
            df = df.append({"coords":loc},ignore_index=True)
    loc = bookmark # Return to bookmarked location

# Reinitializing our search. This time with an offset to fill in the gaps left by the first passs   
begin_new = np.array([lat_f(coords_bk[0], -radius_bk+np.sqrt(3)*radius_hd), # Note: offset(lat) = np.sqrt(3)*radius_hd
                      lng_f(coords_bk[1],coords_bk[0],radius_hd)]) # Note: offset(lng) = radius_hd
loc = begin_new
#print("BEGIN: {}".format(begin_new))
while haversine(loc,coords_bk) < radius_bk:
    bookmark = loc # placing a bookmark so we can return here later
    if haversine(loc,coords_bk) > radius_bk:
        break
    df = df.append({"coords":loc},ignore_index=True)
    while True:
        loc_new = np.array([loc[0],lng_f(loc[1],loc[0],2*radius_hd)])  
        loc = loc_new
        if haversine(loc,coords_bk) > radius_bk:
            break
        else:
            df = df.append({"coords":loc},ignore_index=True)
    loc = bookmark
    while True:
        loc_new = np.array([loc[0],lng_f(loc[1],loc[0],-2*radius_hd)])  
        loc = loc_new
        if haversine(loc,coords_bk) > radius_bk:
            break
        else:
            df = df.append({"coords":loc},ignore_index=True)
    loc=bookmark 
    loc_new = np.array([lat_f(loc[0],2*np.sqrt(3)*radius_hd),loc[1]])
    loc = loc_new          
df.rename_axis("hood_no",inplace=True)
df.shape
```

#### Fetching addresses <a name="fetch-address"></a>

Now that we have our grid of neighborhoods, the coordinates of each can be reverse geocoded to find the associated address. OSM's Nomatim API is used to fetch addresses. The terms of this API propose a limit of no more than **one call per second** and **results must be cached**. Let's be a good-sport and hardcode a delay into the loop.

```python
import time # wait a second between requests to OSM
import pickle # for caching results
```

First a function is written to return an address of a location given its coordinates.

```python
def get_address(nparr_coords):
    lat = str(nparr_coords[0])
    lng = str(nparr_coords[1])
    url = "https://nominatim.openstreetmap.org/reverse?format=json&lat={}&lon={}".format(lat,lng)
    try:
        response = requests.get(url,headers=headers) 
    except requests.exceptions.RequestException as e:
        print("Error: Bad request\n    {}".format(e))
        raise e
    if response.status_code == 200:
        return response.json()['address']
    else:
        print("Warning: HTTP status {}".format(response.status_code))
```

Now we will create a dataframe of addresses. They are loaded from file if possible. Otherwise, the *get_address* function is called to fetch them from OSM.

```python
#!rm brooklyn-ny-15.pickle
```

```python
df_addr = pd.DataFrame(columns=['addr'])
addr_file = '{}-{}-{}.pickle'.format(city,state,str(df.shape[0]))
if os.path.isfile(addr_file):
    print("Loading '{}' from disk...".format(addr_file))
    with open(addr_file,'rb') as f:
        df_addr = pickle.load(f)
    #df_addr = pickle.load(addr_file)
else:
    i=0
    print("Fetching addresses from OSM...")
    for each in df['coords']:
        address = get_address(each)
        os.sys.stdout.write("{} {}\r".format(i,address[list(address.keys())[1]]))
        df_addr = df_addr.append({'addr':address},ignore_index=True)
        time.sleep(1)
        i = i + 1
    with open(addr_file,'wb') as f:
        pickle.dump(df_addr,f)
print('\nDone.')
```

Following, a neighborhood's *competitiveness* is also heuristically determined with the quantity and proximity of the following types of venues:

#### NOTE:
Foursquare limits the number of API calls per day for their unpaid users like me. To limit the number of calls, a single call is made to fetch BOTH the hipster-venues and the competitive-venues for each neighborhood. Afterwards, the venues are processed locally to determine to which category the venue belongs.

Further complicating the matter, when a call is made to Foursquare for a specific category, Foursquare returns venues within the specified category as well as categories which Foursquare believes is similar.

To this end, a list of all categories resulting from an API call for the hipster-venues was created. Each venue is compared to this list to determine its type. The list is stored in the file **hiplike.pickle** which must be loaded to continue with analysis.


```python
# This file contains a list of all categories returned by
# foursquare when a call is made for the Hipster venues below.
with open('hiplike.pickle','rb') as f:
    hiplike = pickle.load(f)

# Hipster category defintions:
organic_grocer = '52f2ab2ebcbc57f1066b8b45'
bicycle_shop = '4bf58dd8d48988d115951735'
flea_market = '4bf58dd8d48988d1f7941735'
indie_movies = '4bf58dd8d48988d17e941735'

hipster_list = [organic_grocer,bicycle_shop,flea_market,indie_movies]
hipster_str = "{},{},{},{}".format(organic_grocer,
                               bicycle_shop,
                               flea_market,
                               indie_movies)

# Competition category defintions:
diner = '4bf58dd8d48988d147941735'
breakfast_spot = '4bf58dd8d48988d143941735'
coffee_shop = '4bf58dd8d48988d1e0931735'
cafe = '4bf58dd8d48988d16d941735'
burger_joint = '4bf58dd8d48988d16c941735'
sandwich_place = '4bf58dd8d48988d1c5941735'
greek_restaurant = '4bf58dd8d48988d10e941735'

competition_list = [diner,breakfast_spot,coffee_shop,cafe,burger_joint,sandwich_place,greek_restaurant]
competition_str = "{},{},{},{},{},{},{}".format(diner,
                                          breakfast_spot,
                                          coffee_shop,
                                          cafe,
                                          burger_joint,
                                          sandwich_place,
                                          greek_restaurant)
```

Defining a function to fetch venues given a set of coordinates. Hipster and competition venues are combined into a single API call in an attempt to not exceed the quota for unpaid users.

Input: list of coordinates
<br/>
Output: JSON object

```python
CLIENT_ID = 'HJDDCUREEXYLVQ1IXBCPT3DYLX41RUIJKGPDUA15FY0WGJ5J'
CLIENT_SECRET = 'KPGNM3TWALV5QBZKKYZY3LMLBSOHLQYH4LIFSE2TLO3HJ35J'
VERSION = '20201208'
LIMIT=50
CATEGORY='{},{}'.format(hipster_str,competition_str)    # Combining venues into a single call
                                                        # to avoid reaching daily quota
def get_venues(coords):
    url='https://api.foursquare.com/v2/venues/search/?v={}&ll={},{}&radius={}&client_id={}&client_secret={}&categoryId={}&limit={}'.format(VERSION,coords[0],coords[1],radius_hd,CLIENT_ID,CLIENT_SECRET,CATEGORY,LIMIT)
    try:
        response = requests.get(url)
        return response.json()['response']['venues']
    except requests.exceptions.RequestException as e:
        raise e
```

Defining a function to parse the JSON object resultant of the API call to Foursquare.

Input: Foursquare JSON response<br/>
Input: The neighborhood number associated with the response<br/>
Output: Pandas DataFrame containing values pertinent to this analysis<br/>

```python
def filter_response(response,index):
    filtered_response = pd.DataFrame(columns=['hood_no','category','name','coords_venue','distance','is_hip'])
    for venue in response:
        is_hip = False                          # Venues where "is_hip = False" are competition
        tmp_list = []
        category_ = ''
        for category in venue['categories']:    # Check if venue categories are in the hip list
            category_ = category['name']
            for hip in hiplike:
                if category['id'] == hip:
                    tmp_list.append(True)
        is_hip = any(tmp_list)                  # Toggle is_hip if any venue categories are hip
        coords_venue = np.array([venue['location']['lat'],venue['location']['lng']])
        distance = venue['location']['distance']
        tmp_dict = {'hood_no':index,
                    'category':category_,
                    'name':venue['name'],
                    'coords_venue':coords_venue,
                    'distance':distance,
                    'is_hip':is_hip}
        filtered_response = filtered_response.append(tmp_dict,ignore_index=True)
    return filtered_response
```

With our two newly-defined functions, code is crafted to create a dataframe containing all of the venues relevant to this analysis. An attempt is made to load venues from file before fetching them from Foursquare.

```python
venuesfile = 'venues-{}-{}-{}.pickle'.format(city,state,df.shape[0])
if os.path.isfile(venuesfile):
    print('Loading: {}'.format(venuesfile))
    with open(venuesfile,'rb') as f:
        venues = pickle.load(f)
else:
    print('fetching venues...')
    venues = pd.DataFrame(columns=['hood_no','category','name','coords_venue','distance','is_hip'])
    for index, coords in zip(df.index, df['coords']):
        try:
            response = get_venues(coords)
            venues = venues.append(filter_response(response,index))
        except:
            continue
    venues.reset_index(inplace=True)
    with open(venuesfile,'wb') as f:
        pickle.dump(venues,f)
        print('Venues dumped to {}'.format(venuesfile))

print('Done')
venues.drop(columns='index',inplace=True)
print(venues.shape)
venues[venues['hood_no'] == 146]
```

#### Scoring Neighborhoods <a name='scoring-neighborhoods'></a>

A sample section of data is displayed below. Sections like this one are used to generate a score for each neighborhood. Neighborhoods are scored according to the criteria defined when we first set off.

* Close to hipster-venues
* Far from competitive-venues
* Close to major infrastructure

```python
df_merge[ (df_merge['hood_no'] == 73) & (df_merge['is_hip'] == True) ]
```

Our neighborhood selection criteria prefers neighborhoods within which hipster-venues are closeby. Unfortunately, the data we have if the opposite of what we want to know. We will use the distance data from Foursquare to calculate how close a venue is to the neighborhood center. 

Let's define the function *Closeness* to calculate how close a venue is to a neighborhood center and normalize it.


<br/>
$D_{max} - maximum\ distance\ to\ any\ venue$

$d_v - distance\ to\ venue\ v$

$C(d_v) -\ Closeness\ of\ venue\ v$

<center><br/>
$C(d_v) = \frac{D_{max} - d}{D_{max}}$

```python
max_distance = venues['distance'].max()

def closeness(distance): # More is better
    return (max_distance - distance)/max_distance   
```

Now we will use our closeness function to generate a proximity score **X** for each neighborhood. The second term of the proximity score is added to represent the number of venues in a neighborhood within the score. For hipster venues, the proximity score will be negative. Competition venues will contribute a postive proximity score.

$X_{H}(d)- \ proximity\ score\ of\ neighborhood\ H$

$d_h -\ distance\ to\ hipster\ venue$

$d_c -\ distance\ to\ competition\ venue$

$\hat{h} -\ hipster\ space$

$\hat{c} -\ competition\ space$

<center><br/>
$\vec{X_{H}} = \sum_{h}(C\left(d_{h}\right) + 1)\hat{h} + \sum_{c}(C\left(d_{c}\right) + 1)\hat{c}$

```python
def proximity_score(hood):
    values = [] #List to hold our vector magnitudes
    for truth in [True,False]:
        part = hood.loc[df_merge['is_hip'] == truth]
        value = sum(part['distance'].apply(closeness))+len(part)
        values.append(value)
    return values
```

Below a funtion to return the shortest distance between a neighborhood and any bridge normalized to the diameter of our search area. Distances are calculated with the Haversine equation defined above.

<br><center>
$ \vec{B_H} = min(\frac{H(\rho_{H},\rho_{b})}{2r_{bk}})\hat{b} $

```python
def bridgedist(coords_hd):
    tmp_list = []
    for bridge in df_bridges['coords']:
        tmp_list.append(haversine(bridge, coords_hd))
    return min(tmp_list)/(2*radius_bk)
```

Finally we are able to assemble all of the pertinent information for each neighborhood. For each neighborhood we will now calculate each component of the score vector **S**.

<center>
$\vec{S_H} = \vec{X_H} + \vec{B_H}$

<center><br/>
$\vec{S_H} = h\hat{h} + c\hat{c} + b\hat{b}$

```python
S = np.zeros(shape=(len(df.index),3))   # This analysis lends itself to numpy.

for hood in df.index:
    tmp_list = []
    selection = df_merge[ (df_merge['hood_no'] == hood)] 
    tmp_list = proximity_score(selection)
    tmp_list.append(bridgedist(df['coords'][hood]))
    S[hood] = tmp_list

c_max =np.amax(S[:,1]) #Let's remember the largest competition magnitude for later
print("Components of the Score vector for a sample of five neighborhoods:")
print(S[0:5])
print("Maximum competition score: {}".format(c_max))
```

Now we make some meaning out what we have calculated. Our end goal is to select neighborhoods that have the lowest score possible. Neighborhoods with a large hipster component should be preferred and so let's take the magnitude of the hipster-component and negate it.
<center><br>
$V_H = -h$

Further, we would like to penalize neighborhoods with lots of competition. Let's add a term to penalize neighborhoods with a lot of competition.
<center><br>
$V_H = -h +\frac{c}{C_{max}}$

Currently, the hipster term is dominating this formula. Let's scale the competition term to the value of the hipster-term. This way, the most competitive neighborhood will negate any value brought by hipster-venues.
<center><br>
$V_H = -h +h\frac{c}{C_{max}}$

Finally, we would will prefer locations that are closer one of the bridges over locations that are further. Let's add a final term for that penalty.
<center><br>
$V_H = -h +h\frac{c}{C_{max}}+(1+b)^2$

Below we define a function to produce a score value V from our score vector **S**. 

```python
def value(vector):
    if vector[0] !=0: # Check if any hipster venues exist
        return -vector[0] + vector[0]*vector[1]/c_max + np.power(1 + vector[2],2)
    else:
        return np.float64(5.0) # Penalizing neighborhoods that do not have any hipster venues
```

Numpy allows us to quickly and efficiently calculate all of the score values from each neighborhood's vector components.

```python
scores = np.apply_along_axis(value,axis=1,arr=S)
print(scores[0:10])
df['hoodscores'] = scores
```

#### Visualizing Scores <a name='visualizing-scores'></a>
Here I define a function to generate a color for each neighborhood depending on its score value.

```python
score_max = df['hoodscores'].sort_values().unique()[-2]
score_min = df['hoodscores'].min()
score_mean = df['hoodscores'].mean()

color = [0,0,0]
magnitude = 255
m = np.pi/(2*(score_max - score_min))

def hood_color(hood_score):
    if hood_score == 5:
        return '#000000'
    else:
        theta = hood_score*m - m*score_min
        color[0] = int(magnitude*np.sin(theta))
        color[1] = int(magnitude*np.cos(theta))
        color[2] = int(0)
        hxstr = '#'
        for c in color:
            hxstr = hxstr + '{0:0{1}X}'.format(c,2)
        return hxstr
```
